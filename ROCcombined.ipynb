Goal: just plot ROC for all three methods together

```python
import numpy as np
import matplotlib.pyplot as plt

auc_tree = 0.675
auc_bay = 0.820
auc_nn = 0.827
tpr_tree = np.load("./data/tpr_tree.npy")
fpr_tree = np.load("./data/fpr_tree.npy")
tpr_bay = np.load("./data/tpr_bay.npy")
fpr_bay = np.load("./data/fpr_bay.npy")
tpr_nn = np.load("./data/tpr_nn.npy")
fpr_nn = np.load("./data/fpr_nn.npy")

plt.figure(figsize=(6, 5))

plt.plot(fpr_bay, tpr_bay, label=f"Approach A: Bayesian Network ROC curve, AUC = {auc_bay:.3f}", linestyle="-")
plt.plot(fpr_tree, tpr_tree, label=f"Approach B: Decision Tree ROC curve, AUC = {auc_tree:.3f}", linestyle=":")
plt.plot(fpr_nn, tpr_nn, label=f"Approach C: Neural Network ROC curve, AUC = {auc_nn:.3f}", linestyle="-.")

plt.plot([0, 1], [0, 1], linestyle="--", label="Random classifier")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
